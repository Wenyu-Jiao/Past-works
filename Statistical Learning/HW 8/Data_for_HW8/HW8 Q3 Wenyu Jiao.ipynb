```python
import pandas as pd
import numpy as np
q3 = pd.read_csv("hepatitis.csv")
y = q3['target'].values
q3.pop('target')
q3.pop('ID')
q3.replace('?', np.nan, inplace = True)
q3 = q3.astype('float64')
x = q3.fillna(q3.mean())
#for i in q3.columns:
#    q3[i] = pd.to_numeric(q3[i],errors='coerce')
#q3 = q3.fillna(q3.mean())
#print(q3.columns)
x.head()
```

```python
from sklearn import svm
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

cols = ['age', 'bili', 'alk', 'sgot', 'albu', 'protime']
rest = x[x.columns.difference(cols)].columns

sd1 = sc_X.fit_transform(x[cols])
col_pd = pd.DataFrame(sd1, columns = cols)

X = np.concatenate((col_pd, x[rest]), axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle = False)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
regressor = SVC()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
print("y_pred",y_pred)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.001,0.1,1, 10],'gamma' : [0.1, 0.01, 1, 10],'kernel': ['rbf', 'poly', 'sigmoid','linear']}
grid2 = GridSearchCV(SVC(), param_grid) 

grid2.fit(X_train, y_train) 
# print best parameter after tuning
print(grid2.best_params_)
grid_predictions2 = grid2.predict(X_test)
print(classification_report(y_test, grid_predictions2)) 
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, grid_predictions2))
```
