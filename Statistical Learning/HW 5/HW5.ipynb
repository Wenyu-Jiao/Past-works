```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.random.uniform(0,1,(100,1))*5-2
mu, sigma = 0, 1
s = np.random.normal(mu, sigma, (100,1))
Y = 0.5*x*x+0.5*x+1+s
plt.plot(x,Y,'bx')
xsort = np.sort(x)
line = np.array(np.sort(x,axis = None)).reshape(100,1)
poly = PolynomialFeatures(1, include_bias=False)
tran_X = poly.fit_transform(xsort)
for i in [0, 10, 100]:
    clf = Ridge(alpha=i,solver = 'sag')
    reg = clf.fit(tran_X, Y)
    line_poly = poly.transform(line)
    y_hat = reg.predict(line_poly)
    plt.plot(line, y_hat, label = "alpha="+str(i))
plt.xlabel("x")
plt.ylabel("y")
plt.title("ridge regression with sag")
plt.legend()
plt.show()
```

```python
plt.plot(x,Y,'bx')
for i in [0, 10, 100]:
    clf = Ridge(alpha=i,solver = 'svd')
    reg = clf.fit(tran_X, Y)
    line_poly = poly.transform(line)
    y_hat = reg.predict(line_poly)
    plt.plot(line, y_hat, label = "alpha="+str(i))
plt.xlabel("x")
plt.ylabel("y")
plt.title("ridge regression with svd")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x,Y,'bx')
xsort = np.sort(x)
line = np.array(np.sort(x,axis = None)).reshape(100,1)
poly = PolynomialFeatures(10, include_bias=False)
tran_X = poly.fit_transform(xsort)
alpharig = [0, 10**(-5), 1]
alpharig_beta = []
for i in alpharig:
    clf = Ridge(alpha=i, solver = 'sag')
    reg = clf.fit(tran_X, Y)
    line_poly = poly.transform(line)
    y_hat = reg.predict(line_poly)
    alpharig_beta.append(reg.coef_)
    plt.plot(line, y_hat, label = "alpha="+str(i))
plt.xlabel("x")
plt.ylabel("y")
plt.title("ridge regression with p = 10 with solver sag")
plt.legend()
plt.show()
## alpha = 0 is poly regression
```

```python
plt.plot(x,Y,'bx')
for i in alpharig:
    clf = Ridge(alpha=i, solver = 'svd')
    reg = clf.fit(tran_X, Y)
    line_poly = poly.transform(line)
    y_hat = reg.predict(line_poly)
    plt.plot(line, y_hat, label = "alpha="+str(i))
plt.xlabel("x")
plt.ylabel("y")
plt.title("ridge regression with p = 10 with solver svd")
plt.legend()
plt.show()
## alpha = 0 is poly regression
```
